# Plots

## Type

```python
plt.figure(figsize=(5, 4))
groups = train_df.groupby(by='type', axis=0).count()['item']
plt.xticks(rotation = 45, ha = 'right')
plt.bar(x=groups.index, height=groups.values)
plt.title(label = 'Distribution of the feature type in the training dataset')
plt.tight_layout()
```

```python
type_label_counts = train_df.groupby(['type', 'label']).size().unstack()


# Plot
type_label_counts.plot(kind='bar', stacked=True, figsize=(7, 5), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Distribution of the Feature Type by Label")
plt.xlabel("Type")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Categories

```python
# List of all the categories
category_cols = [
    'category_architecture', 'category_biology', 'category_comics and anime',
    'category_fashion', 'category_films', 'category_food', 'category_geography',
    'category_literature', 'category_media', 'category_music',
    'category_politics', 'category_sports', 'category_transportation',
    'category_visual arts', 'category_combined'
]

# Sum the categories, as it's 0 or 1
category_counts = train_df[category_cols].sum().sort_values(ascending=False)

# Visualisation
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title("Category frequence")
plt.xlabel("Category")
plt.ylabel("Number of occurence")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sum over the labels, for every category
cat_label_counts = train_df.groupby(['label'])[category_cols].sum().T

# Plot
cat_label_counts.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Category Frequency by Label")
plt.xlabel("Category")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Date

```python
# Set parameters
bin_size = 50

# Create bins
min_year = train_df['date'].min()
max_year = train_df['date'].max()
bins = range(min_year, max_year + bin_size, bin_size)
labels = [f"{start}-{start + bin_size - 1}" for start in bins[:-1]]

# Cut years into bins
train_df['year_binned'] = pd.cut(train_df['date'], bins=bins, labels=labels, right=False)

# Count frequencies
freq_by_bin = train_df['year_binned'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12, 6))
freq_by_bin.plot(kind='bar')
plt.title(f"Frequency of date")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
freq_by_bin_label = train_df.groupby(['year_binned', 'label']).size().unstack(fill_value=0)

# Plot
freq_by_bin_label.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Frequency of Items by Year Range and Label")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Nb_lang

```python
plt.figure(figsize=(5, 4))
groups = train_df.groupby(by='nb_lang', axis=0).count()['item']
plt.xticks(rotation = 45, ha = 'right')
plt.bar(x=groups.index, height=groups.values)
plt.title(label = 'Distribution of the feature nb_lang in the training dataset')
plt.tight_layout()
plt.show()
```

```python
# Parameters
bin_size = 20

# get range
min_val = int(train_df['nb_lang'].min())
max_val = int(train_df['nb_lang'].max())
bins_range = range(min_val, max_val + bin_size, bin_size)
bin_labels = [f"{i}-{i + bin_size - 1}" for i in bins_range[:-1]]

# create bin column
train_df['lang_bin'] = pd.cut(train_df['nb_lang'], bins=bins_range, labels=bin_labels, right=False)

# count by bin and label
bin_counts = train_df.groupby(['lang_bin', 'label']).size().unstack().fillna(0)

# plot
bin_counts.plot(kind='bar', stacked=True, figsize=(9, 5), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("nb_lang distribution by label")
plt.xlabel("nb_lang (binned)")
plt.ylabel("count")
plt.xticks(rotation=45)
plt.legend(title="label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Correlation



```python
# Selection of the variables to examine for correlation.
var_to_corr = ['subclass_depth',
       'subclasses', 'instances_of', 'instances_of_up', 'date', 'nb_lang', 'descr_num_nouns', 'descr_num_verbs',
       'descr_num_adjectives', 'descr_has_location', 'descr_has_ethnic_group',
       'descr_has_event', 'type_cat', 'category_architecture',
       'category_biology', 'category_comics and anime', 'category_fashion',
       'category_films', 'category_food', 'category_geography',
       'category_literature', 'category_media', 'category_music',
       'category_politics', 'category_sports', 'category_transportation',
       'category_visual arts', 'category_combined',
       'main_country_cat']

# Compute the correlation
corr = train_df[var_to_corr].corr()

# Plot
plt.figure(figsize=(12, 9))
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
