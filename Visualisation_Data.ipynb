## Date

```python
# Set parameters
bin_size = 50

# Create bins
min_year = train_df['date'].min()
max_year = train_df['date'].max()
bins = range(min_year, max_year + bin_size, bin_size)
labels = [f"{start}-{start + bin_size - 1}" for start in bins[:-1]]
print(labels)

# Cut years into bins
train_df['year_binned'] = pd.cut(train_df['date'], bins=bins, labels=labels, right=False)

# Count frequencies
freq_by_bin = train_df['year_binned'].value_counts().sort_index()

# we get rid of the negligible intervals with less than 10 records
freq_by_bin = freq_by_bin[freq_by_bin  >60]

freq_by_bin

# Plot
plt.figure(figsize=(12, 6))
freq_by_bin.plot(kind='bar')
plt.title(f"Frequency of date")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
freq_by_bin_label = train_df.groupby(['year_binned', 'label']).size().unstack(fill_value=0)

freq_by_bin_label = freq_by_bin_label[freq_by_bin_label.sum(axis=1) >60]


# Plot
freq_by_bin_label.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Frequency of Items by Year Range and Label")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Date dev

```python
# Set parameters
bin_size = 50

# Create bins
min_year = dev_df['date'].min()
max_year = dev_df['date'].max()
bins = range(min_year, max_year + bin_size, bin_size)
labels = [f"{start}-{start + bin_size - 1}" for start in bins[:-1]]
print(labels)

# Cut years into bins
dev_df['year_binned'] = pd.cut(dev_df['date'], bins=bins, labels=labels, right=False)

# Count frequencies
freq_by_bin = dev_df['year_binned'].value_counts().sort_index()

# we get rid of the negligible intervals with less than 10 records
freq_by_bin = freq_by_bin[freq_by_bin  >3]

freq_by_bin

# Plot
plt.figure(figsize=(12, 6))
freq_by_bin.plot(kind='bar')
plt.title(f"Frequency of date")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
freq_by_bin_label = dev_df.groupby(['year_binned', 'label']).size().unstack(fill_value=0)

freq_by_bin_label = freq_by_bin_label[freq_by_bin_label.sum(axis=1) >3]


# Plot
freq_by_bin_label.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Frequency of Items by Year Range and Label")
plt.xlabel("Year Range")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
