## Date

```python
# Set parameters
bin_size = 50

# Create bins
min_year = train_df['date'].min()
max_year = train_df['date'].max()
bins = range(min_year, max_year + bin_size, bin_size)
labels = [f"{start}-{start + bin_size - 1}" for start in bins[:-1]]


# Cut years into bins
train_df['year_binned'] = pd.cut(train_df['date'], bins=bins, labels=labels, right=False)

# Count frequencies
freq_by_bin = train_df['year_binned'].value_counts().sort_index()

# we get rid of the negligible intervals with less than 10 records
freq_by_bin = freq_by_bin[freq_by_bin  >60]

freq_by_bin

# Plot
plt.figure(figsize=(12, 6))
freq_by_bin.plot(kind='bar')
plt.title(f"frequency of date, the intervals with ath least 1% of records,\n train set")
plt.xlabel("Year range")
plt.ylabel("number of occurrences")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
freq_by_bin_label = train_df.groupby(['year_binned', 'label']).size().unstack(fill_value=0)

freq_by_bin_label = freq_by_bin_label[freq_by_bin_label.sum(axis=1) >60]


# Plot
freq_by_bin_label.plot(kind='bar', stacked=True, figsize=(14, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("distribution of labels grouped by the interval of occurence,\n train set ")
plt.xlabel("year range")
plt.ylabel("number of occurrences")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

###  distribution of labels with respect to the existance of the date information

```python
freq_by_bin_label = train_df.groupby(['year_binned', 'label']).size().unstack(fill_value=0)

freq_binary = train_df.copy()

indices = freq_binary['date'] != -1
freq_binary.loc[indices, 'date'] =1
freq_binary.loc[~indices,'date'] =0



props = freq_binary.groupby(['date', 'label']).size().unstack(fill_value=0)

props.plot(kind='bar', stacked=True, figsize=(10,6), color=['#1f77b4', '#ff7f0e', '#2ca02c'])


plt.title("distribution of labels grouped by existence of date info, \ntrain set")
plt.xlabel("data existence")
plt.ylabel("count of occurences")
plt.xticks(ticks=[0, 1], labels=['not-exists', 'exists'], rotation=0)
plt.legend(title="label", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
