```python
from tensorflow.keras.models import Model, load_model
model_logits_description = load_model("model_logits_description.keras")
model_logits_features = load_model("model_logits_features.keras")
```

# Export models to HuggingFace

```python
from huggingface_hub import login, create_repo, upload_folder
import os
import tensorflow as tf
from huggingface_hub import hf_hub_download

# Login to HuggingFace
login(token="")

# 1. Create a public repo on Hugging Face under your own space
repo_name = "MNLP-HW1"
repo_id = f"tgarnier067/{repo_name}"

# Create the repo if it doesn't already exist
create_repo(repo_id, exist_ok=True, private=False)

# 2. Prepare a folder to upload multiple models
model_folder = "models"

# Create a folder to place all the models
os.makedirs(model_folder, exist_ok=True)

# 3. Save multiple models and move them to this folder
models = {
    "model_logits_description": model_logits_description,
    "model_logits_features": model_logits_features
}

# Save the models in separate files
for model_name, model in models.items():
    model_file_path = os.path.join(model_folder, f"{model_name}.keras")
    model.save(model_file_path)

# 4. Add a default README
readme_content = """# MNLP HW1 Models
This repository contains multiple models trained for different tasks. Each model is saved as a separate `.keras` file.
"""

# Create a README.md file in the model folder
with open(os.path.join(model_folder, "README.md"), "w") as f:
    f.write(readme_content)

# 5. Upload all models to the Hugging Face repo
upload_folder(
    repo_id=repo_id,
    folder_path=model_folder,
    path_in_repo=""
)

print(f"All models have been successfully uploaded to Hugging Face in the repo {repo_id}!")
```

# Import models from HuggingFace

```python
# Repo name and model file name
repo_id = "tgarnier067/MNLP-HW1"
model_name = "model_logits_features.keras"

# Download the model from Hugging Face
model_path = hf_hub_download(repo_id=repo_id, filename=model_name)

# Load the .keras model with TensorFlow
model = tf.keras.models.load_model(model_path)

# Use the model for predictions or other tasks
model.summary()
```
